### Import independencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
```

### Converting data csv files to Dataframes

```python
APD_csv = pd.read_csv("Crime_Reports_2017.csv")
results_2017_df = APD_csv
census_csv=pd.read_csv('Median_Income.csv', delimiter=';')
```

### Merging census income data with Austin Crime reports

```python
df=pd.merge(results_2017_df,census_csv, how='left', left_on='Zip Code', right_on='ZIP')
df
```

```python
df.count()
```

### Monthly Crime Analysis

```python
#adding month column to the data
df['month'] = pd.DatetimeIndex(results_2017_df['Report Date']).month
df
```

```python
Incidents_monthly=df.groupby('month').agg({'Incident Number':'count'})
Incidents_monthly
grouped_monthly=Incidents_monthly.reset_index()
grouped_monthly
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
plt.bar(months, grouped_monthly['Incident Number'], align="center" )
plt.xlabel("Month")
plt.ylabel("Number of Reports")
plt.title("Monthly Crime Rates 2017")
plt.tight_layout()
plt.show()
```

```python
grouped_offense=df.groupby(['Highest Offense Description']).agg({'Incident Number':'count'})
grouped_offense.sort_values('Incident Number', ascending=False, inplace=True)
grouped_offense
```

```python
grouped_zip=df.groupby('ZIP').agg({'Incident Number':'count', 'Median_Income':'mean' })
grouped_zip
grouped_zip.sort_values('Incident Number', ascending=False)
```
